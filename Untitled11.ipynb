```python
def impute_brs(cols):
  total_bedrooms=cols[0]
  households=cols[1]

  if pd.isnull(total_bedrooms):
    return households+ 40
  else:
    return total_bedrooms



df["total_bedrooms"] = df[["total_bedrooms","households"]].apply(impute_brs,axis=1)
```

```python
df.drop(["ocean_proximity"],axis=1,inplace=True)
```

```python
df.head()
```

```python
objc=["ocean_proximity"]
for colname in objc:
  df[colname]=df[colname].astype('category')
```

```python
df.select_dtypes(['category']).columns
```

```python
OP=pd.get_dummies(df['ocean_proximity'],drop_first=True)
```

```python
df=pd.concat([df,OP],axis=1)
df.head()
```

```python
y=df['median_house_value']
```

```python
x=df.drop('median_house_value',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.20, random_state=1001)
```

```python
df.head()
```

```python
from sklearn.linear_model import LogisticRegression

lreg= LogisticRegression()

lreg.fit(x_train,y_train)
```

```python
y_predict_lreg = lreg.predict(x_test)
print(y_predict_lreg)
```

```python
from sklearn.metrics import classification_report,accuracy_score

print("Classification Model")
print("--"*30)

logre_accuracy=round(accuracy_score(y_test,y_predict_lreg)*100,2)

print("Accuracy = ",logre_accuracy,"%")
```

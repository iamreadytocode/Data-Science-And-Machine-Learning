1.Write a program that takes a string as input and counts the number of vowels and
consonants in the string. (5 marks)
Input: Hello, World!
Output:
Number of vowels: 3
Number of consonants: 7


```python
string= input("Enter the string ")
vowels=0
consonents=0

for letter in string:
  if(letter.isalpha()):

    if(letter=='a' or letter=='e' or letter=='i' or letter=='o' or letter=='u' or letter=='A' or letter=='E' or letter=='I' or letter=='O' or letter=='U'):
      vowels=vowels+1
    else:
      consonents=consonents+1

print("The vowels are = ",vowels)
print("The consonents are = ",consonents)
```

5. Write a Python program that calculates the Body Mass Index (BMI) for a person
based on their weight (in kilograms) and height (in meters). The BMI is calculated
using the following formula: (5 marks)
BMI = (weight) / (height^2)

Instructions:

● Prompt the user to enter their weight in kilograms.
● Prompt the user to enter their height in meters.
● Calculate the BMI using the provided formula.
● Display the calculated BMI to the user.

Additionally, provide an interpretation of the BMI according to the
following categories:

● BMI < 18.5: Underweight
● 18.5 <= BMI < 25: Normal weight
● 25 <= BMI < 30: Overweight
● BMI >= 30: Obese


```python
weight=int(input("Enter your weight in kilograms"))
height= float(input("Enter your height in meters"))
bmi=weight/(height*height)

if bmi<18.5:
  print("You  are  underweight: ",bmi)
elif bmi>=18.5 and bmi<25:
  print("Your weight is normal: ",bmi)
elif bmi>=25 and bmi<30:
  print("You are overweight: ",bmi)
elif bmi>=30:
  print("You are obeeeeseeeee: ",bmi)
```

9. Create a list containing the 10 numbers . Print the first Three prime numbers from
the list. (5 marks)
Expected Output: List of the first three prime numbers: [2, 3, 5]

```python
def prime(n):
  if n<=1:
    return False
  elif n<=3:
    return True
  elif n%2==0 or n%3==0:
    return False
  i = 5
  while i*i<n:
    if n%i==0 or n%(i+2)==0:
      return False
    i+=6
  return True


ml=[6,7,9,11,12,13,15,19,20,21]
nl=[]
count=0
for item in ml:
  if prime(item)==True:
    count+=1
    if count<4:
      nl.append(item)
print(nl)
```

13.Write a program that takes two dictionaries as input and merges them into a new
dictionary. If there are common keys, combine their values. (5 marks)
dict1= {‘a’:5, ‘b’:10, ‘c’:15}
dict2= {‘b’:20, ‘c’:25, ‘d’:30}

Output: {'a': 5, 'b': 30, 'c': 40, 'd': 30}

```python
dict1= {'a':5, 'b':10, 'c':15}
dict2= {'b':20, 'c':25, 'd':30}
new_dict={}
for key,value in dict1.items():
    if key not in new_dict:
      new_dict[key]=value
    else:
      new_dict[key]+=value
for key1,value1 in dict2.items():
    if key1 not in new_dict:
      new_dict[key1]=value1
    else:
      new_dict[key1]+=value1

print(new_dict)
```

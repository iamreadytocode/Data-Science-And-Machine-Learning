```python
from sklearn import preprocessing

y = y.reshape(-1,1)

prepro2= preprocessing.StandardScaler().fit(y)

y_trans = prepro2.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_trans,y,test_size=0.20,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

lr = LogisticRegression()

lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
print(y_pred.shape)
print(y_pred)
```

```python
sns.scatterplot(x=y_test,y=y_pred,color='red',label="Actual data points")
plt.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)],color="green",label="Ideal line(Model)")
plt.legend()
plt.show()
```

```python
result= np.column_stack((y_test,y_pred))

print("Actual values      Predicted values")

print("--"*30)

for actual,predicted in result:
  print(f"{actual:6.2f} | {predicted:14.2f}")
```

```python
from sklearn.metrics import accuracy_score

y_pred = lr.predict(x_test)

print("Logistic Regression")

print("--"*30)

accuracy= round(accuracy_score(y_test,y_pred)*100,2)

print("Accuracy = ",accuracy,"%")
```

```python
x=dd.drop("Price",axis=1)
y=dd["Price"]
```

```python
from sklearn import preprocessing

y = y.values.reshape(-1, 1)

pp1=preprocessing.StandardScaler().fit(y)

y_trans=pp1.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_trans,y_trans,test_size=.20,random_state=101)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

linreg= LinearRegression()

linreg.fit(x_train,y_train)
```

```python
y_pred=linreg.predict(x_test)
print(y_pred.shape)
print(y_pred)
```

```python
sns.scatterplot(x=y_test,y=y_pred,color='purple',label='Actual Data Points')
plt.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)],color='red',label="Ideal Line")
plt.legend()
plt.show()
```

```python
result=np.column_stack((y_test,y_pred))

print("  Actual Value | Predicted Value")

print("--"*30)

for actual,predicted in result:
  print(f"{actual:14.2f} | {predicted:12.2f}")
```

```python
residual= actual- y_pred.reshape(-1)
print(residual)
```

```python
sns.distplot(residual,kde=True)
```

```python
from sklearn.metrics import mean_squared_error

print("Linear Regression Model")
print("--"*30)

mse= mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)

print("The mean square error is",mse)
print("The root mean square error is",rmse)
```

```python

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

rfg= DecisionTreeRegressor()
rfg.fit(x_train,y_train)

y_pred_rf= rfg.predict(x_test)

DTr = mean_squared_error(y_pred_rf,y_test)

print("Decission Tree Regression",DTr)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

rfg= RandomForestRegressor()
rfg.fit(x_train,y_train)

y_pred_rf= rfg.predict(x_test)

RFr = mean_squared_error(y_pred_rf,y_test)

print("Decission Tree Regression",RFr)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

rfg= GradientBoostingRegressor()
rfg.fit(x_train,y_train)

y_pred_rf= rfg.predict(x_test)

GBr = mean_squared_error(y_pred_rf,y_test)

print("Decission Tree Regression",GBr)
```

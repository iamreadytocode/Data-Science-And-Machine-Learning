```python
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
%matplotlib inline
import matplotlib.pyplot as plt
import sklearn
warnings.filterwarnings("ignore")
plt.rcParams["figure.figsize"]=[10,5]
```

```python
def impute_age(cols):
  Age=cols[0]
  Pclass=cols[1]
  if pd.isnull(Age):
    if Pclass ==1:
      return 34
    elif Pclass==2:
      return 29
    else:
      return 24
  else:
    return Age


df["Age"]=df[["Age","Pclass"]].apply(impute_age,axis=1)
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
df.drop(["Cabin"],axis=1,inplace=True)
```

```python
df.head(1)
```

```python
objc=["Sex"]
for colname in objc:
  df[colname]=df[colname].astype('category')
```

```python
df.select_dtypes(['category']).columns
```

```python
OP=pd.get_dummies(df["Sex"],drop_first=True)
```

```python
df=pd.concat([df,OP],axis=1)
df.head()
```

```python
y=df["Survived"]
```

```python
x=df.drop('Survived',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.20, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression

lreg= LogisticRegression()

lreg.fit(x_train,y_train)
```

```python
y_predict_lreg = lreg.predict(x_test)
print(y_predict_lreg)
```

```python
from sklearn.metrics import classification_report,accuracy_score

print("Classification Model")
print("--"*30)

logre_accuracy=round(accuracy_score(y_test,y_predict_lreg)*100,2)

print("Accuracy = ",logre_accuracy,"%")
```

```python
from sklearn.tree import DecisionTreeClassifier

print("Decision Tree classifier")

dtre=DecisionTreeClassifier()

dtre.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score

y_pred_dtre= dtre.predict(x_test)

print("Decision Tree classifier")
print("--"*30)

dtre_accuracy=round(accuracy_score(y_test,y_pred_dtre)*100,2)

print("Accuracy = ",dtre_accuracy,"%")
```

```python
from sklearn.ensemble import RandomForestClassifier

print("RandomForestClassifier")

rf=RandomForestClassifier()

rf.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score

y_pred_rf= rf.predict(x_test)

print("RandomForestClassifier")
print("--"*30)

rf_accuracy=round(accuracy_score(y_test,y_pred_rf)*100,2)

print("Accuracy = ",rf_accuracy,"%")
```

```python
from sklearn.ensemble import GradientBoostingClassifier

print("GradientBoostingClassifier")

grc=GradientBoostingClassifier()

grc.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score

y_pred_grc= grc.predict(x_test)

print("GradientBoostingClassifier")
print("--"*30)

grc_accuracy=round(accuracy_score(y_test,y_pred_grc)*100,2)

print("Accuracy = ",grc_accuracy,"%")
```
